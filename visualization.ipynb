# Visualization of Data Extraction

```python
import fitz

from main import get_dataframe_from_table, get_dataframes_from_doc, find_tables_from_page
from utils import draw_cells, show_image
```

## Visualization of the Extracted Table using PyMuPDF Package

```python
doc_path = "./Mercedes-Benz-EQS.pdf"

# Open the PDF document using PyMuPDF
doc = fitz.open(doc_path)

# Choose a page for visualization (e.g., page 7)
page_number = 7
page = doc[page_number - 1]

# Find tables on the selected page
tables = find_tables_from_page(page)

# Visualize the extracted tables on the page
for table in tables:
    draw_cells(page, table)
show_image(page)
```

## Convert the table to a Pandas DataFrame

```python
table = tables[0]
table.to_pandas()
```

## Get extracted and formatted dataframe from the selected table with the constraints

```python
get_dataframe_from_table(table)
```
